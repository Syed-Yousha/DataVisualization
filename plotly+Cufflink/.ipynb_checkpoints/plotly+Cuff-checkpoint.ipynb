```python
# Generating a Random DataFrame
df = pd.DataFrame(np.random.randn(100,4), columns='A B C D'.split())
```

```python
df.head()
```

```python
df2 = pd.DataFrame({'Category':['A', 'B', 'C'], 'Values':[32,42,56]})
```

```python
df2
```

```python
%matplotlib inline #for displaying matplot graphs
```

```python
df.plot() #builtin python plot which use matplot, line plot
```

```python
df.iplot() #interactive plotly line plot
```

```python
df.iplot(kind='scatter', x='A', y='B', mode = 'markers') #scatter plot, kind is used to describe plotting
```

```python
df2
```

```python
df2.iplot(kind='bar', x='Category', y='Values')
```

```python
df.iplot(kind='bar') #as there are many rows so the plot will be messy
```

```python
df.mean().iplot(kind='bar') # we use aggragate function to make it readable
```

```python
df.iplot(kind= 'box') #box plot
```

```python
df3 = pd.DataFrame({'x':[1,2,3,4,5], 'y':[11,22,33,44,55], 'z':[5,4,3,2,1]})
df3
```

```python
df3.iplot(kind='surface',colorscale='rdylbu') #srface plot with red yellow blue colorscale
```

```python
df['A'].iplot(kind='hist',bins=55)
```

```python
df[['A','B']].iplot(kind='spread')
```

```python
#Bubble plot
df2.iplot(kind='bubble', x='A',y='B')
```
