```python
import numpy as np
import seaborn as sns
import pandas as pd
%matplotlib inline
```

```python
df1 = pd.read_csv('df1', index_col=0)
df1.head()
```

```python
df1['A'].hist(bins=30)
```

```python
#Area plot  [df2 is categorical data]
df2 = pd.read_csv('df2')
df2.head()
```

```python
df2.plot.area() #.plot.area will be used to plot area plot
```

```python
df2.plot.area(alpha=0.4)
```

```python
#barplot panda visualizations are used for quick graphical code
df2.plot.bar()
```

```python
df2.plot.bar(stacked=True) #attribute to to stack the bars
```

```python
#histogram
df2.plot.hist(bins=50) #bins are bars
```

```python
#line plot, used for sequential series as df1 which have a time sequence index
df1.plot.line(y='B') #as x is defined, which is orginial index of df1 we only have to give y parameter
```

```python
#scatter plot
df1.plot.scatter(x='A', y='B') #here we have to give x and y 
```

```python
df1.plot.scatter(x='A', y='B', c='C') #parameter and attribute c is color, here is 3 dimensional scatter plot
```

```python
df1.plot.scatter(x='A', y='B', s=df1['C']*50) #we can also add 3rd parameter with size
```

```python
#box plot for categorical data
df2.plot.box()
```

```python
#hex plot for variant random data, unstructured data

df = pd.DataFrame(np.random.randn(1000,3), columns = ['a', 'b', 'c'])
df.head()
```

```python
df.plot.hexbin(x='a',y='c', gridsize=25) #gridsize is used for size
```

```python
df.plot.hexbin(x='a',y='c', gridsize=25, cmap='inferno') #can also use cmaps colors
```

```python
#kde plot
df.plot.kde()
```

```python
df['a'].plot.kde() #can also be used for 1 column
```
