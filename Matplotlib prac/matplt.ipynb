```python
import numpy as np
x = np.linspace(0, 5, 11)
y = x**2
x
```

```python
y
```

```python
#Functional method
plt.plot(x,y)
```

```python
plt.plot(x,y)
plt.show()
```

```python
#Labeling and coloring
plt.plot(x,y, 'r-')
plt.xlabel('Time')
plt.ylabel('Displacement')
plt.title('Velocity')
```

```python
#multiple plots
plt.subplot(1,2,1) #(row, col, pos)
plt.plot(x,y,'g')

plt.subplot(1,2,2)
plt.plot(y,x,'y')
plt.show() #not much impportant in jupyter notebook
```

```python
#object Oriented Method, use to gain more ccontrol over plot
fig = plt.figure() #plain canvas
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8]) #axes are used for movement and enlargement of the figure
axes.plot(x, y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Title')
```

```python
#Playing with axes

fig = plt.figure()

axes1 = fig.add_axes([0.1, 0.1, 0.8,0.8])
axes2 = fig.add_axes([0.2, 0.5, 0.4,0.3])

axes1.plot(x,y)
axes1.set_title('Main Plot')
axes2.plot(y,x)
axes2.set_title('Sub Plot')
```

```python
#Subplotting again
fig, axes = plt.subplots(nrows = 1, ncols = 2)
axes[0].plot(x,y,'r') 
axes[0].set_title('Plot 1')

axes[1].plot(y,x,'g')
axes[1].set_title('Plot 2')
```

```python
fig, axes = plt.subplots(nrows = 3, ncols = 3)
plt.tight_layout() #for good layout
```

```python
#Figure size and DPI

fig = plt.figure(figsize=(4,2)) #figsize = (width, height) , 

ax = fig.add_axes([0,0,1,1]) #add_axes(horizontal move, vertical move, width, height)  Max is 1
ax.plot(x,y)
```

```python
#Subplots and figsizes

fig, axes = plt.subplots(nrows = 2, ncols = 1,figsize = (6,3))
axes[0].plot(x,y)
axes[1].plot(y,x, 'r')

plt.tight_layout()
```

```python
fig.savefig('my_pic.png', dpi = 200)
```

```python
#legends

figr = plt.figure()
axes =  figr.add_axes([0,0,1,1])
axes.plot(x**2,y,'r',label = 'X sqr')
axes.plot(x**3,y,'g',label = 'X cube')
#axes.legend(loc=(0.5, 0.8)) #for custom legend location
axes.legend(loc=0) # We have to put labels on plots to make legend work. loc = 0  is for best location for legend
```

```python
# -------------------------------------------- PLOT APPEARANCE -------------------------------------------- 
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])


#ax.plot(x,y, 'r') # one word color
#ax.plot(x,y, 'orange') # whole word color
#ax.plot(x,y, '#FF8') # hex code custom color

#ax.plot(x,y, 'r', lw = 1) # lw => line width
ax.plot(x,y, color= 'r', lw = 5 , alpha = 0.6 ,ls = ':') #alpha is for transparency and linestyle is for style
```

```python
fig = plt.figure()
axe = fig.add_axes([0,0,1,1])
axe.plot(y,x, marker = '*', markersize = 10) #Marker is used to mark main data points and marker size is used to manipulate their size
axe.plot(x*2,y, marker = 'o', markersize = 15, markerfacecolor = 'yellow', markeredgewidth='2', markeredgecolor='blue')
```

```python
#Limits

fig = plt.figure()
axe = fig.add_axes([0,0,1,1])
axe.plot(x,y)

axe.set_xlim([0,2]) #X limite
axe.set_ylim([0,3]) #Y limite
```
