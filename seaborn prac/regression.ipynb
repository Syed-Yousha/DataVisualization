```python
import seaborn as sns
%matplotlib inline
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex') #linear model plot
```

```python
    sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex', markers = ['o','v']) #used for  markers shape
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex', markers = ['o','v'], scatter_kws = {'s':100}) #scatter_kws is used for increasong size
```

```python
    sns.lmplot(x='total_bill', y='tip', data=tips, col='time' ,hue='sex') #col can be use to factor the plot
```
