```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
tips = sns.load_dataset('tips')
tips.head()
```

```python
#styling
sns.set_style('whitegrid')
sns.countplot(x='sex', data=tips)
sns.despine() #use to remove borders and spines
```

```python
#figsize can be use to change size of plots
plt.figure(figsize= (12,5))
sns.countplot(x='smoker', data=tips)
```

```python
#context and scale
sns.set_context('poster', font_scale =1) #seaborn assume that we have to put the plot on a poster.
sns.countplot(x='time', data=tips)
```

```python
#changing colors
sns.set_context('notebook')
sns.lmplot(x='total_bill', y='tip', data=tips, hue='time', palette='inferno') #palette setring are on google, hue if used for categoriess
```
