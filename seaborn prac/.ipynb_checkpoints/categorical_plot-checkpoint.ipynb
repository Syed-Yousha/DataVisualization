```python
import seaborn as sns
import numpy as np
%matplotlib inline
tips = sns.load_dataset('tips')
tips.head()
```

```python
import numpy as np
```

```python
#barplot: use aggregate function, and catagorical data on x-axis
sns.barplot(x='sex' , y='total_bill', data=tips, estimator=np.std)
```

```python
#Count Plot: Will count the number of occurences of x-axis, which will be category column
sns.countplot(x='sex', data=tips)
```

```python
#Box Plot: show categorical data through boxes
sns.boxplot(x= 'day', y='tip', data=tips)
```

```python
sns.boxplot(x= 'day', y='tip', data=tips, hue='smoker') #hue is used to add another data layer to current plot
```

```python
#Violin Plot: almost same as box plot but also show kde and a have one more argument
sns.violinplot(x='day', y='total_bill', data=tips)
```

```python
sns.violinplot(x='day', y='total_bill', data=tips, hue='smoker')
```

```python
sns.violinplot(x='day', y='total_bill', data=tips, hue='smoker', split=True)
```

```python
#strip Plot: a scatter plot based on categories
sns.stripplot(x='day',y='total_bill',data=tips)
```

```python
sns.stripplot(x='day',y='total_bill',data=tips, hue= 'sex')
```

```python
#swarm plot:mix of strip plot and violin plot, not so much popular
sns.swarmplot(x='day',y='total_bill',data=tips)
```

```python
#we can combine it with violin plot to make them more readable
sns.violinplot(x='day',y='total_bill',data=tips)
sns.swarmplot(x='day',y='total_bill',data=tips, color='blue')
```
