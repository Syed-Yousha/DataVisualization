```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
iris = sns.load_dataset('iris')
iris.head()
```

```python
#pair plot, shows all column x rows in grid format
sns.pairplot(iris)
```

```python
#more customization of pairplot using pairGrid

p = sns.PairGrid(iris)
p
```

```python
# Map different plots to different parts of the grid
p = sns.PairGrid(iris)

p.map_diag(sns.histplot)  # Diagonal - histogram
p.map_upper(sns.scatterplot)  # Upper triangle - scatterplot
p.map_lower(sns.kdeplot)  # Lower triangle - KDE
```

```python
#Facet grid 

tips = sns.load_dataset('tips')
tips.head()
```

```python
g = sns.FacetGrid(data=tips,col='time', row='smoker')
```

```python
g = sns.FacetGrid(data=tips,col='time', row='smoker') #grids are accordance with time and smoker
g.map(sns.distplot, 'total_bill') #X-axis will be total_bill, Y-axis will be density 
```

```python
g = sns.FacetGrid(data=tips,col='time', row='smoker')
g.map(plt.scatter, 'total_bill', 'tip') #X-axis will be total_bill, Y-axis will be tip 
```
