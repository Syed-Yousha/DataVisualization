```python
import seaborn as sns
%matplotlib inline

tips = sns.load_dataset('tips')
flights = sns.load_dataset('flights')
tips.head()
```

```python
flights.head()
```

```python
#heatmap: make use of matrix dataset and plot like heated pixels
fp=flights.pivot_table(index='month', columns='year', values='passengers') # pivot_table is used to make relation with index and cols
fp
```

```python
sns.heatmap(fp)
```

```python
#some other attributes
sns.heatmap(fp, cmap='coolwarm', linecolor='black', linewidth=1)
```

```python
#cluster maps: arrange the data with concentrated clusters
sns.clustermap(fp)
```

```python
#some more attributes to change color and scale, it can become more understandable
sns.clustermap(fp, cmap='coolwarm', standard_scale=1) #warm(red) means more flghts and vice versa
```
