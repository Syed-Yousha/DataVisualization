```python
#seaborn Distribution
import seaborn as sns
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.distplot(tips['total_bill'], kde = False, bins = 30)
```

```python
sns.displot(tips['total_bill'], kde = False, bins = 50)
```

```python
#joint plot
sns.jointplot(x='total_bill',y='tip',data=tips)
```

```python
#kind parameter addition Hex
sns.jointplot(x='total_bill',y='tip',data=tips, kind = 'hex')
```

```python
#kind parameter addition regression
sns.jointplot(x='total_bill',y='tip',data=tips, kind = 'reg')
```

```python
#kind parameter addition kde (kernal density estimation)
sns.jointplot(x='total_bill',y='tip',data=tips, kind = 'kde')
```

```python
#pair plot (will give you visualization for all variables{rows x cols})
sns.pairplot(tips)
```

```python
#With Hue argument: we can catogarise our plot based on binary cols like gender
sns.pairplot(tips, hue='sex') 
```

```python
#Rug plot
sns.rugplot(tips['total_bill'])
```

```python
#Kde plot
sns.kdeplot(tips['tip'])
```
